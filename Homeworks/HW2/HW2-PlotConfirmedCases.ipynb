```python
df_us = df[df['Country/Region'] == 'US']
df_us = df_us.set_index('Country/Region', drop = True) # set pandas dataframe index to country
df_us = df_us.drop(columns=['Lat','Long','Province/State'])

data_us = df_us.loc['US','2/28/20':'4/16/20']
data_us
```

## 2a: Plot Italy and US data in two axes

Use subplots and assign separate colors to each country. Set the same limits on the y-axis for both subplots

```python
fig,ax = plt.subplots(1,2,figsize=(20,5)) # or whatever dimensions you like
ax[0].plot(data_us.index,data_us.values,'.-',color='r',label='United States')
ax[1].plot(data_italy.index,data_italy.values,'.-',color='b',label='Italy')
ax[0].set_title('US Confirmed Cases')
ax[1].set_title('Italy Confirmed Cases')
for a in ax:
    a.set_xlabel('calendar data')
    a.set_ylabel('Confirmed COVID-19 Cases')
    for tick in a.get_xticklabels():
        tick.set_rotation(90)
plt.show()
```

## 2b: Plot Italy and US data in the same plot

Use the same colors as above, and include a legend

```python
fig,ax = plt.subplots(1,1,figsize=(20,5)) # or whatever dimensions you like
ax.plot(data_us.index,data_us.values,'-',color='r',linewidth=3,label='United States')
ax.plot(data_italy.index,data_italy.values,'-',color='b',linewidth=3,label='Italy')
plt.xticks(rotation=90) # rotate the xticks so text is not so tight
plt.xlabel('date')
plt.ylabel('Confirmed COVID-19 cases')
plt.legend()
plt.show()
```

## 2c Plot Italy and US data on log axis

Note that we only want the y-axis to be logarithmic

```python
fig,ax = plt.subplots(1,1,figsize=(20,5)) # or whatever dimensions you like
ax.plot(data_us.index,data_us.values,'-',color='r',linewidth=3,label='United States')
ax.plot(data_italy.index,data_italy.values,'-',color='b',linewidth=3,label='Italy')
plt.yscale('log')
plt.ylim(bottom=1)
plt.grid(which='major')
plt.xticks(rotation=90) # rotate the xticks so text is not so tight
plt.xlabel('date')
plt.ylabel('Confirmed COVID-19 cases')
plt.legend()
plt.show()
```

## 2d Make the same plot as above, but using a `for` loop

```python
data_list = [data_us,data_italy]
data_names = ['US','Italy']

fig,ax = plt.subplots(1,1,figsize=(12,8)) # or whatever dimensions you like

for data,name in zip(data_list,data_names):
    ax.plot(data.index,data.values,'-',linewidth=3,label=name)
plt.yscale('log')
plt.ylim(bottom=1)
plt.grid(which='major')
plt.xticks(rotation=90) # rotate the xticks so text is not so tight
plt.xlabel('Date')
plt.ylabel('Confirmed COVID-19 cases')
plt.legend()
plt.show()
```

```python
fig,ax = plt.subplots(1,1,figsize=(12,8)) # or whatever dimensions you like

country_names = ['US','China','Italy','Singapore','United Kingdom']

for name in country_names:
    df_t = df[df['Country/Region'] == name]
    df_t = df_t.set_index('Country/Region', drop = True) # set pandas dataframe index to country
    df_t = df_t.drop(columns=['Lat','Long','Province/State'])
    
    data = df_t.loc[name,'2/1/20':'4/15/20']
    if data.ndim>1:
        data=data.sum()
        ax.plot(data,'-',linewidth=3,label=name)
    else:
        ax.plot(data.index,data.values,'-',linewidth=3,label=name)

plt.yscale('log')
# plt.ylim(bottom=1)
plt.grid(which='major')
plt.xticks(rotation=90) # rotate the xticks so text is not so tight
plt.xlabel('Date')
plt.ylabel('Confirmed COVID-19 cases')
plt.legend()
plt.show()
```
